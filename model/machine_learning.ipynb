```python
X_GROUP = [Xgroup0,Xgroup1,Xgroup2,Xgroup3,Xgroup4,Xgroup5,Xgroup6,Xgroup7,Xgroup8,Xgroup9]
Y_GROUP = [ygroup0,ygroup1,ygroup2,ygroup3,ygroup4,ygroup5,ygroup6,ygroup7,ygroup8,ygroup9]
```

```python
acc_of_group = {}
var_of_group = {}
count = 0
for X,Y in zip(X_GROUP,Y_GROUP) :
    
    print('Group_{}'.format(count))
    
    
    #### 標準化
    
    X = minmax.fit_transform(X)
    
    print(X.shape,'原來維度')
    
    
    #### 降維度  使用extratree
    
    
    X = SelectFromModel(ExtraTreesRegressor(n_jobs=-1).fit(X,Y),prefit=True).transform(X)
    
    print(X.shape,'下降後維度')
    
    
    #### 帶入模型
    
    ### forest ##############
    start = time.time()

    acc = []
    acc.append(cross_val_score(forest, X, Y, cv=kf, n_jobs = -1))
    print('randomforest',acc)
    print('randomforest_mean',np.array(acc).mean())
    print('randomforest_std',np.array(acc).std())
    
    acc_of_group['forest_group{}'.format(count)] = np.array(acc).mean()
    var_of_group['forest_group{}'.format(count)] = np.array(acc).std()
    
    end = time.time()
    print('forest_runtime',end - start)

    
    ### extra_tree #############
    start = time.time()

    acc = []
    acc.append(cross_val_score(extra, X, Y, cv=kf, n_jobs = -1))
    print('extree',acc)
    print('extree_mean',np.array(acc).mean())
    print('extree_std',np.array(acc).std())

    acc_of_group['extra_group{}'.format(count)] = np.array(acc).mean()
    var_of_group['extra_group{}'.format(count)] = np.array(acc).std() 
    
    end = time.time()
    print('extree_runtime',end - start)


    ### xgboost ################
    start = time.time()

    acc = []
    acc.append(cross_val_score(xgbreg, X, Y, cv=kf, n_jobs = -1))
    print('xgb',acc)
    print('xgb_mean',np.array(acc).mean())
    print('xgb_std',np.array(acc).std())

    acc_of_group['xgb_group{}'.format(count)] = np.array(acc).mean()
    var_of_group['xgb_group{}'.format(count)] = np.array(acc).std()
    
    end = time.time()
    print('xgb_runtime',end - start)


    ### lightxgboost ###################
    start = time.time()

    acc = []
    acc.append(cross_val_score(lightreg, X, Y, cv=kf, n_jobs = -1))
    print('lightxgb',acc)
    print('lightxgb_mean',np.array(acc).mean())
    print('lightxgb_std',np.array(acc).std())

    acc_of_group['light_group{}'.format(count)] = np.array(acc).mean()
    var_of_group['light_group{}'.format(count)] = np.array(acc).std()
    
    end = time.time()
    print('lightxgb_runtime',end - start)
    
    count +=1
    
    
    
    
    print('=======================================================','Finish')
```

# 找出每個group中表現最好的Model 並且加以儲存

```python
count_a = 1
model_a = []
mean_acc_a = []
select_group_model_acc = {}
for model,mean_acc in zip(acc_of_group,acc_of_group.values()) :
    
    if count_a % 4 != 0 :
        
        model_a.append(model)
        mean_acc_a.append(mean_acc)
    
    
    ## 走一回 GROUP
    if count_a % 4 == 0 :
        
        model_a.append(model)
        mean_acc_a.append(mean_acc)    
    
    
        acc_index = mean_acc_a.index(max(mean_acc_a))

        select_group_model_acc[model_a[acc_index]] = max(mean_acc_a)

        # 清空 List

        model_a = []
        mean_acc_a = []
    
    count_a += 1
```

```python
count_v = 1
model_v = []
mean_var_v = []
select_group_model_var = {}

for model,mean_var in zip(var_of_group,var_of_group.values()) :
    
    if count_v % 4 != 0 :
        
        model_v.append(model)
        mean_var_v.append(mean_var)
    
    
    ## 走一回 GROUP
    if count_v % 4 == 0 :
        
        model_v.append(model)
        mean_var_v.append(mean_var)  
    
    
        var_index = mean_var_v.index(min(mean_var_v))

        select_group_model_var[model_v[var_index]] = min(mean_var_v)

        # 清空 List

        model_v = []
        mean_var_v = []
    
    count_v += 1
```
